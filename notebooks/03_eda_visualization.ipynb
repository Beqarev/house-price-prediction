# Exploratory Data Analysis and Visualization

This notebook generates the 5 required visualizations using the visualization module.

## Step 1: Import Libraries and Load Processed Data

This cell imports the visualization functions from our custom module and loads the processed dataset that was created in the previous notebook.

```python
import pandas as pd
import sys
import matplotlib.pyplot as plt

# Add src to path
sys.path.append('../src')

from visualization import (
    plot_price_distribution,
    plot_correlation_heatmap,
    plot_price_vs_income,
    plot_price_by_age_group,
    plot_pair_plot
)

# Load processed data
df = pd.read_csv('../data/processed/california_housing_processed.csv')
print(f"Dataset shape: {df.shape}")
```

## Visualization 1: Price Distribution (Histogram with KDE)

This visualization shows the distribution of house prices using a histogram with a Kernel Density Estimation (KDE) curve overlay. It helps understand the shape and spread of the target variable.

```python
plot_price_distribution(df, target_col='MedHouseVal', output_dir='../reports/figures')
```

## Visualization 2: Correlation Heatmap

This heatmap shows the correlation coefficients between all numerical features. It helps identify which features are strongly related to each other and to the target variable (house price).

```python
plot_correlation_heatmap(df, output_dir='../reports/figures')
```

## Visualization 3: Price vs. Median Income (Scatter Plot with Regression Line)

This scatter plot shows the relationship between median income and house prices, with a regression trend line. It helps visualize the strength and direction of the relationship between income and housing prices.

```python
plot_price_vs_income(df, price_col='MedHouseVal', income_col='MedInc', output_dir='../reports/figures')
```

## Visualization 4: Price Distribution by House Age Group (Box Plot)

This box plot shows the distribution of house prices grouped by house age categories. It helps identify if house age has an impact on price and shows the variability within each age group.

```python
plot_price_by_age_group(df, price_col='MedHouseVal', age_col='HouseAge', output_dir='../reports/figures')
```

## Visualization 5: Pair Plot (Multivariate Relationships)

This pair plot shows pairwise relationships between key variables (Price, Income, Average Rooms, and House Age). It provides a comprehensive view of how multiple variables relate to each other simultaneously.

```python
plot_pair_plot(df, columns=['MedHouseVal', 'MedInc', 'AveRooms', 'HouseAge'], output_dir='../reports/figures')
```
